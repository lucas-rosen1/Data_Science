```python
## Exercise 4
## Write you own aggregation function and use it after doing groupby on any column (EMP is allowed)
## Tell me what this is doing, what the analysis means, and why its interesting
data = pd.read_csv('data/drugs.tsv',delimiter='\t',index_col='QUESTID2')
emp_dict = {1:'Full Time',2:'Part Time',3:'Unemployed',4:'Other',99:'Child'}
data.EMP = data.EMP.apply(lambda x: emp_dict[x])
def square(ser):
    return np.sum(ser)
data.groupby('EMP').apply(square)

#This is summing up the total number who has ever done any drug for each age group.
#This is interesting because we can see interesting trends such that it appears the data was gathered the most from full time workers and they utilize drugs the most/
#It is interesting to see how the most common drug utilized in general is alcohol, next is cigs, next is MJ, and lastly coc.
#This is interesting because it follows the relationship of most critical crime to least critical.
```
